```python
import numpy as np
import pandas as pd
```

```python
pd.Series(index=['h','e','y'],data=[10,20,30])
```

In pandas, a Series is a 1-dimensional labeled array.
Think of it like:
* a single column in Excel
* or a NumPy array with labels (indexes)

Creating a series in multiple ways
1. Lists
2. Numpy Array
3. Dictionary
4. Custom Index
5. Scaler value

```python
# Series from List
alpha = ['Abfg','bjkyuk','Cgukjfg','Dukug']
pd.Series(alpha)
```

```python
batsman_run= [35,46,24,57,46,68,46]
pd.Series(batsman_run,dtype='int32')
```

```python
subject_name = ['maths' , 'science', 'english', 'computer']
subject_marks = [85,82,91,95]
pd.Series(index=subject_name,data=subject_marks)
```

```python
# creating series from numpy array
NArray = np.random.randint(1,100,10)
print(NArray)
pd.Series(NArray)
```

```python
# Dictionary to Series
data_items = {'Maths' : 85,'English' : 78 , 'Science' : 89, 'Computer' : 89}
pd.Series(data_items)
```

```python
# custom index to series
Values=[1,2,3,4,5]
pd.Series(data=Values,index=['a','b','c','d','e'])
```

```python
# Scaler Value inside an Series
pd.Series(data=4,index=['a','b','c'])
```
