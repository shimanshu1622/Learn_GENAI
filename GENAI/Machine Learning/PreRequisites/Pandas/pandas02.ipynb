### Mathematical Fucntions in Series and Indexing in Series

```python
import numpy as np
import pandas as pd
```

```python
path = r"D:\pythonProjectGenAI\Learn_GENAI\GENAI\DataSets\youtube_dataset\youtube_channel_info_v1.csv"
```

```python
df = pd.read_csv(path)

df.head()
```

```python
subs = df['subscriber_count']
subs
```

```python
#mathematical functions
# 1. sum() -> sum of all values present in series
subs.sum()
```

```python
# 2. product() -> product of all values present in series 
subs.product()
```

```python
subs.sort_values()
```

```python
df.head()
```

```python
# 3. mean() -> avg of the series or dataset column
subs.mean()
```

```python
# 4. Meadian() -> sort the series in asending order the middle one is meadian, used when there is genrally outliers like in salary some have 7-10lpa and some few have 40-50lpa but it will increase the mean(avg) so use meadian 
subs.median()
```

```python
# 5. Mode() -> which value getting repeated maximum times 
subs.mode()
```

```python
#6. std() -> if i need to get the idea that how the values inside the series sre distrubuted from the average or mean vale then we cancome up with the standard deviation
# salary -> 7,8,6,6.5,7.5   all are near to the 7  ( mean/avg) so -> low std
# salary -> 7,15,4,20,28,17 high std bcz from the mean/avg the values are high and low not nearby
subs.std()
```

```python
# 7. variance -> it also provides the insight about the distribution of data from the mean vale and variance is the square of std
# standard_deviation = |mean - v_1| + |mean - v_2|....../ total count of vales
# variance = std**2
subs.var()
```
