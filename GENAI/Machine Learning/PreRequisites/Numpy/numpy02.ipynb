### Mathematical operations 

```python
# Scaler Operation -> a scaler value operation will be done with the array or a numeric value ( 1,4 ....)
array = np.random.rand(3,3)*100
array
```

```python
array = np.round(array)
array
```

```python
array = array.astype('int32')
array
```

```python
np.round(np.random.rand(3,3) * 100).astype('int32') # in one line 
```

```python
# Scaler operations with out matrix or array
print(array)
array+2  # add, sub, mul, divide, floor, modulo 
```

```python
# relational operations
print(array)
array>60
```

```python
# Vector operations on two given matrix 
arr1 = np.arange(1,17).reshape(4,4)
arr2 = np.arange(17,33).reshape(4,4)

print('arr1')
print(arr1)
print('arr2')
print(arr2)
print('arr1 + arr2')  # ->  + - * / 
print(arr1 + arr2)
```

### Important functions

```python
# max() -> max value from that matric , array
np.max(arr1)
```

```python
# axis --> help in getting the max or min value from specific row or column
# axis=0 column wise operation
# axis=1 row wise operation
array
```

```python
np.max(array,axis=0) # max value of each column
```

```python
np.max(array,axis=1) # max value from each row 
```

```python
print(np.min(array,axis=0)) # min value of each column
print(np.min(array,axis=1)) # min value of each row
```

```python
# sum() -> calculates the total sum
array
```

```python
print(np.sum(array)) # TOTAL SUM
print(np.sum(array,axis=0)) # TOTAL SUM COLUMN WISE
print(np.sum(array,axis=1)) # TOTAL SUM ROW WISE
```

```python
# prod() -> product of all values  
print(np.prod(array)) # TOTAL PRODUCT
print(np.prod(array,axis=0)) # TOTAL PRODUCT COLUMN WISE
print(np.prod(array,axis=1)) # TOTAL PRODUCT ROW WISE
```
