### Statistical functions in numpy

```python
import numpy as np
```

```python
array = np.arange(1,10).reshape(3,3)
array
```

```python
# mean() -> mean of all value inside the array or matrix
print(np.mean(array)) # output in -> float64
print(np.mean(array, axis=0)) # column wise
print(np.mean(array, axis=1)) # row wise
```

```python
# std() -> standerd deviation 
print(np.std(array)) # output in -> float64
print(np.std(array, axis=0)) # column wise
print(np.std(array, axis=1)) # row wise
```

```python
# meadian() -> calculates the meadian
print(np.median(array)) # output in -> float64
print(np.median(array, axis=0)) # column wise
print(np.median(array, axis=1)) # row wise
```

```python
# var() -> variance of all values
print(np.var(array)) # output in -> float64
print(np.var(array, axis=0)) # column wise
print(np.var(array, axis=1)) # row wise
```
