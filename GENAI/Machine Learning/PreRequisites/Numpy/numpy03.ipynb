### Statistical functions in numpy

```python
import numpy as np
```

```python
array = np.arange(1,10).reshape(3,3)
array
```

```python
# mean() -> mean of all value inside the array or matrix
print(np.mean(array)) # output in -> float64
print(np.mean(array, axis=0)) # column wise
print(np.mean(array, axis=1)) # row wise
```

```python
# std() -> standerd deviation 
print(np.std(array)) # output in -> float64
print(np.std(array, axis=0)) # column wise
print(np.std(array, axis=1)) # row wise
```

```python
# meadian() -> calculates the meadian
print(np.median(array)) # output in -> float64
print(np.median(array, axis=0)) # column wise
print(np.median(array, axis=1)) # row wise
```

```python
# var() -> variance of all values
print(np.var(array)) # output in -> float64
print(np.var(array, axis=0)) # column wise
print(np.var(array, axis=1)) # row wise
```

### Trignometric functions

```python
 # sin() tan() cos()
np.sin(0)
np.tan(45)
```

```python
#dot product --> matrix1[m x n] * matrix2[p x q]  {n == p}  -->  output[m x q]
# np.dot(m1,m2)

arr1 = np.random.randint(1,10,(4,3))
arr2 = np.random.randint(1,10,(3,5))
print('arr1')
print(arr1)
print('arr2')
print(arr2)
#DOT PRODUCT
dotprod = np.dot(arr1,arr2) # [4 x 3] * [3 x 5] -> [4 x 5]
print(dotprod)
```

```python
# log() -> log of all values of matrix
np.log(arr1)
```

```python
#exp() -> exponential value inside matrix
np.exp(arr2)
```

```python
# round() -> round the float value to nearest possible integer value in float
print(np.round(3.4)) #3.0
print(np.round(3.5)) #4.0
print(np.round(3.8)) #4.0
# ceil() 
print(np.ceil(3.4)) #4.0
print(np.ceil(3.5)) #4.0
print(np.ceil(3.8)) #4.0
# floor()
print(np.floor(3.4)) #3.0
print(np.floor(3.5)) #3.0
print(np.floor(3.8)) #3.0
```

```python
# transpose() => m x n ---> n x m
# np.transpose(arr1)  or arr1.T
print("4 x 3 -->")
print(arr1)
print("3 x 4 -->")
print(np.transpose(arr1))
print("3 x 4 -->")
print(arr1.T)
```

```python
# ravel() -> convert n-d array in 1-d array
print(arr1)
print(np.ravel(arr1))
```

### Indexing and Slicing

```python
oneDArray = np.random.randint(1,10,9)
print(oneDArray)
# target 6th element  array[target-1]
print(oneDArray[5])
```

```python
# 2-D array -> [row,column]
print(arr1) # target 2nd row and 3rd column  -> 8
print(arr1[1,2])
```

```python
# 3-D array -> [z-axis , x-asix , y-axis]
ThreeDArray = np.arange(1,9).reshape(2,2,2)
ThreeDArray
```

```python
# taeget 7 
ThreeDArray[1,1,0]
```

```python
# slicing
# 1-d -> [:]
# 2-d -> [row1:rown,col1:coln]
print(arr1) # taeget middle two rows
print(arr1[1:3])
#target midlle of two elements 1 ,2
print(arr1[1:3,1:2])
```

```python
# Iteration 
# 1-d 
print(oneDArray)
```

```python
for i in oneDArray:
    print(i)
```

```python
for i in oneDArray:
    print(i>5)
```

```python
print(arr1)
for i in arr1:
    print(i)
```

```python
print(ThreeDArray)
for i in ThreeDArray:
    print(i)
```

```python
# nditer() -> iteration over evry value in n-d array
for i in np.nditer(ThreeDArray):
    print(i)
```
