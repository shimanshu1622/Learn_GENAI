```python
np.random.seed(42)
print(np.random.randint(1,10,5))
```

seed() sets the starting point for a pseudo-random number generator (PRNG).

Computers use mathematical formulas to generate numbers that only appear random.
Same seed → same sequence of random numbers.
Different seed → different sequence.

Example:

np.random.seed(42)

42 is just a popular programmer choice from The Hitchhiker's Guide to the Galaxy and has no special mathematical meaning.

```python
#modern way
rng = np.random.default_rng()

rng.random((3,3))
```

```python
rng = np.random.default_rng(43)
rng.random((3,3))
```

Important Difference in Functions

Old:
* np.random.randint()
* np.random.rand()
* np.random.randn()

Modern:
* rng.integers()
* rng.random()
* rng.normal()
