Task 2: Important Mathematical Operations

Given two arrays:

A = np.array([10, 20, 30, 40])

B = np.array([1, 2, 3, 4])

Perform and print results for:

1. Addition (A + B)

2. Subtraction (A - B)

3. Multiplication ( A ^ * B)

4. Division (A / B)

5. Power (A ^ 2)

```python
import numpy as np
```

```python
A = np.array([10,20,30,40])
B = np.array([1,2,3,4])
```

```python
print("A + B")
ArraySum1 = A + B
print(ArraySum1)
```

```python
print("A + B")
ArraySum2 = np.add(A,B)
print(ArraySum2)
```

```python
print("A - B")
ArraySubtracted = np.subtract(A,B)
print(ArraySubtracted)
```

```python
print("A * B")
ArrayProducts = A * B
print(ArrayProducts)
```

```python
print("A / B")
ArrayDivision = A + B
print(ArrayDivision)
```

```python
print("A ** B")
ArrayPower = A ** B
print(ArrayPower)
```
