Task 1: Creating NumPy Arrays

1. Import Numpy as np.

2. Create the following arrays:
* A 1D array of integers from 1 to 10
* A 2D array of shape (3, 3) with values from 1 to 9
* A NumPy array from the list: [10, 20, 30, 40, 50]

3. Print:
* Shape of each array
* Data type of each array

```python
import numpy as np
```

```python
array = np.arange(1,11)
print("1D array ",array)
```

```python
TwoDArray = np.arange(1,10).reshape(3,3)
print("2 D array ", TwoDArray)
```

```python
demolist = [10,20,30,40,50]
Narray = np.array(demolist)
print("numpy array from list",Narray)
```

```python
print(array.shape)
print(array.dtype)
```

```python
print(TwoDArray.shape)
print(TwoDArray.dtype)
```

```python
print(Narray.shape)
print(Narray.dtype)
```
